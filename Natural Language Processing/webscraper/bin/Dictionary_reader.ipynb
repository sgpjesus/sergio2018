```python
with open('ptdict.txt', encoding='utf-8') as f:
    content = f.readlines()
```

```python
content
```

```python
import re
import pandas as pd
```

```python
token_list = []
for element in content:
    token = (re.search('([A-Za-z0-9À-ÿ-]+)', str(element)))
    token_list.append(token[0])
```

```python
cat_list = []
for element in content:
    token = re.search('CAT=[a-z]+', str(element))
    if token:
        cat_list.append(token[0][4:])
    else:
        cat_list.append('Null Category')
```
